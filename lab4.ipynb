## –û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ –∏ —Ä–µ–∞–ª–∏–∑—É–π—Ç–µ –∫—Ä–µ–∞—Ç–∏–≤–Ω—ã–µ –º–µ—Ç–æ–¥—ã –æ—á–∏—Å—Ç–∫–∏ –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö.

re.sub(pattern, repl, string, count=0, flags=0)

pattern: –†–µ–≥—É–ª—è—Ä–Ω–æ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –Ω—É–∂–Ω–æ –Ω–∞–π—Ç–∏ –≤ —Å—Ç—Ä–æ–∫–µ.

repl: –°—Ç—Ä–æ–∫–∞, –Ω–∞ –∫–æ—Ç–æ—Ä—É—é –Ω—É–∂–Ω–æ –∑–∞–º–µ–Ω–∏—Ç—å –Ω–∞–π–¥–µ–Ω–Ω—ã–π —Ñ—Ä–∞–≥–º–µ–Ω—Ç.

string: –ò—Å—Ö–æ–¥–Ω–∞—è —Å—Ç—Ä–æ–∫–∞, –≤ –∫–æ—Ç–æ—Ä–æ–π –±—É–¥–µ—Ç –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç—å –ø–æ–∏—Å–∫ –∏ –∑–∞–º–µ–Ω–∞.

count (–Ω–µ–æ–±—è–∑–∞—Ç–µ–ª—å–Ω—ã–π): –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–º–µ–Ω (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é –∑–∞–º–µ–Ω—è–µ—Ç –≤—Å–µ –≤—Ö–æ–∂–¥–µ–Ω–∏—è).

flags (–Ω–µ–æ–±—è–∑–∞—Ç–µ–ª—å–Ω—ã–π): –§–ª–∞–≥–∏ –¥–ª—è –∏–∑–º–µ–Ω–µ–Ω–∏—è –ø–æ–≤–µ–¥–µ–Ω–∏—è —Ä–µ–≥—É–ª—è—Ä–Ω–æ–≥–æ –≤—ã—Ä–∞–∂–µ–Ω–∏—è (–Ω–∞–ø—Ä–∏–º–µ—Ä, re.IGNORECASE –¥–ª—è –∏–≥–Ω–æ—Ä–∏—Ä–æ–≤–∞–Ω–∏—è —Ä–µ–≥–∏—Å—Ç—Ä–∞).

```python
import re
from nltk.corpus import stopwords

stop_words = set(stopwords.words('russian'))
words_to_remove = {'–Ω–µ', '–≤—Å–µ–≥–¥–∞', '–ª—É—á—à–µ', '–Ω–∏–∫–æ–≥–¥–∞', '—Ö–æ—Ä–æ—à–æ', '–Ω–µ—Ç', '–Ω–µ–ª—å–∑—è', '–º–æ–∂–Ω–æ'}
for word in words_to_remove:
    stop_words.discard(word)

def clean_text(text):
    text = text.lower()
    
    text = re.sub(r'@[\w-]+', '', text) # —É–±–∏—Ä–∞–µ–º —Å—Å—ã–ª–∫–∏ –Ω–∞ –ª—é–¥–µ–π. —Å—á–∏—Ç–∞–µ–º —á—Ç–æ –Ω–∏–∫–∏ —Å–æ—Å—Ç–æ—è—Ç –∏–∑ –±—É–∫–≤, —Ü–∏—Ñ—Ä, –Ω–∏–∂–Ω–∏—Ö –ø–æ–¥—á–µ—Ä–∫–∏–≤–∞–Ω–∏–π –∏ –¥–µ—Ñ–∏—Å–æ–≤
    
    text = re.sub(r'[^\w\s]', '', text) # —É–±–∏—Ä–∞–µ–º –≤—Å–µ —á—Ç–æ –Ω–µ –±—É–∫–≤—ã. ^ = –Ω–µ

    text = re.sub(r'\s+', ' ', text).strip() # —É–±–∏—Ä–∞–µ–º –ª–∏—à–Ω–∏–µ –ø—Ä–æ–±–µ–ª—ã

    tokens = [word for word in text.split() if word not in stop_words] # stopwords
    
    return tokens
```

```python
df.loc[:, 'clean_text'] = df['text'].apply(clean_text)
df
```

## –û—Å—É—â–µ—Å—Ç–≤–∏—Ç–µ —Å—Ç–µ–º–º–∏–Ω–≥ –ø–æ–¥–≥–æ—Ç–æ–≤–ª–µ–Ω–Ω–æ–≥–æ –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö –∏ –ø—Ä–µ–æ–±—Ä–∞–∑—É–π—Ç–µ –∫–∞–∂–¥—ã–π —Ç–≤–∏—Ç –≤ –º–µ—à–æ–∫ —Å–ª–æ–≤.

```python
from nltk.stem import SnowballStemmer

snowball = SnowballStemmer(language="russian")

def stemm(words):
    stemmed = []
    for word in words:
        stemmed.append(snowball.stem(word))
    return stemmed

df.loc[:, 'stemmed_text'] = df['clean_text'].apply(stemm)
df
```

```python
from sklearn.model_selection import train_test_split

X = df['stemmed_text'].apply(lambda x: ' '.join(x)) # count_vectorizer –ø–æ–ª—É—á–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ —Å—Ç—Ä–æ–∫—É
y = df['label'].map({'negative': 0, 'positive': 1})

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 32, stratify=y)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Ç–µ–∫—Å—Ç –≤ –º–µ—à–æ–∫ —Å–ª–æ–≤
X_train_vectorized = vectorizer.fit_transform(X_train)

print(vectorizer.get_feature_names_out())

print(X_train_vectorized.toarray())
X_train_vectorized.toarray()[0].shape
```

```python
X_test_vectorized = vectorizer.transform(X_test)
print(X_test_vectorized.toarray())
X_test_vectorized.toarray()[0].shape
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_vectorized)
X_test_tfidf = tfidf_transformer.transform(X_test_vectorized)

X_train_tfidf.toarray()
```

```python
import spacy

rus_words = spacy.load("ru_core_news_sm")


def lemma(words):
    text = " ".join(words)
    doc = rus_words(text)
    lemmas = [token.lemma_ for token in doc]
    return lemmas

df.loc[:, 'lemmed_text'] = df['clean_text'].apply(lemma)
df
```
