### Разделение данных на обучающую (85%) и тестовую часть (15%)

```python
import pandas as pd
df = pd.read_excel('lr2.xlsx').dropna()
df
```

```python
X = df.drop(columns=['ОКЭР (регион), данные на 2013 год', 'Ybin']) 
y = df['Ybin'] 
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=12, stratify=y)
# stratify = y гарантирует, что распределение y в тренировочной и тестовой выборке будет аналогично исходному распределению
```

## Нормирование исходных данных

```python
from sklearn.preprocessing import StandardScaler

standard_scaler = StandardScaler()

X_train = pd.DataFrame(standard_scaler.fit_transform(X_train))
X_test = pd.DataFrame(standard_scaler.transform(X_test))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(data=X_train)
plt.title("X_train")
plt.show()
sns.boxplot(data=X_test)
plt.title("X_test")
plt.show()
```

## fit-predict модели kNN

параметры GridSearchCV
- estimator: Модель, для которой производится подбор гиперпараметров (например, KNeighborsClassifier()).
- param_grid: Сетка гиперпараметров. Это словарь, где ключи — названия параметров, а значения — списки значений для перебора.
- cv: Количество фолдов для кросс-валидации. Например, cv=5 будет использовать 5-кратную кросс-валидацию.
- scoring: Метрика для оценки качества модели (например, 'accuracy', 'f1', 'roc_auc').
- n_jobs: Количество параллельных процессов. Если n_jobs=-1, используется максимальное количество ядер.

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

param_grid = {'n_neighbors': range(1, 11)}
```

```python
from sklearn.model_selection import GridSearchCV

grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

best_n_neighbors = grid_search.best_params_['n_neighbors']
best_n_neighbors
```

y_train : 12 единиц, 27 нулей. классы несбалансированы, поэтому лучше взять метрику f1 или roc_auc

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold

stratified_kfold = StratifiedKFold(n_splits=5)
grid_search = GridSearchCV(knn, param_grid, cv=stratified_kfold, scoring='f1')
grid_search.fit(X_train, y_train)

best_n_neighbors = grid_search.best_params_['n_neighbors']
best_n_neighbors
```

```python
grid_search = GridSearchCV(knn, param_grid, cv=stratified_kfold, scoring='roc_auc')
grid_search.fit(X_train, y_train)

best_n_neighbors = grid_search.best_params_['n_neighbors']
best_n_neighbors
```

```python
# Обучение модели с наилучшим числом соседей
best_knn = KNeighborsClassifier(n_neighbors=best_n_neighbors)
best_knn.fit(X_train, y_train)
```

## fit-predict модели логистической регрессии

параметры LogisticRegression в scikit-learn:

- penalty: Тип регуляризации, применяемой к модели. 
- C: Обратный коэффициент регуляризации. Определяет степень регуляризации. Меньшее значение C усиливает регуляризацию, предотвращая переобучение. Большое значение C ослабляет регуляризацию, позволяя модели лучше подстроиться под данные.
- solver: Алгоритм, используемый для оптимизации
- max_iter: Максимальное количество итераций алгоритма оптимизации. 
- class_weight: Веса классов для компенсации несбалансированности. 'balanced' — автоматически корректирует веса на основе частоты классов.
Пользователь может задать свои веса для каждого класса в виде словаря.
- multi_class: Определяет, как обрабатывать многоклассовую классификацию:
'ovr' (по умолчанию) — стратегия "один против всех".
'multinomial' — прямая многоклассовая логистическая регрессия (используется с solver='lbfgs', 'sag', 'newton-cg' или 'saga').
- random_state: Фиксирует начальное состояние генератора случайных чисел для воспроизводимости результатов.

```python
from sklearn.linear_model import LogisticRegression
import numpy as np

log_reg = LogisticRegression(max_iter=100)

param_grid = {'C':  np.arange(0.01, 101, 0.1)}

stratified_kfold = StratifiedKFold(n_splits=5)
grid_search = GridSearchCV(log_reg, param_grid, cv=stratified_kfold, scoring='f1')

grid_search.fit(X_train, y_train)

best_C = grid_search.best_params_['C']
grid_search
```

```python
best_log_reg = LogisticRegression(C=best_C, max_iter=100)
best_log_reg.fit(X_train, y_train)
```

### fit-predict модели случайного леса

```python
from sklearn.ensemble import RandomForestClassifier

rand_forest = RandomForestClassifier(random_state=10)

param_grid = {
    'n_estimators': range(1, 151, 5),  # Количество деревьев
    'max_depth': range(1, 14)   # Глубина деревьев
}

stratified_kfold = StratifiedKFold(n_splits=5)    
grid_search = GridSearchCV(rand_forest, param_grid, cv=stratified_kfold, scoring='f1')

grid_search.fit(X_train, y_train)

best_depth = grid_search.best_params_['max_depth']
best_n_estimators = grid_search.best_params_['n_estimators']
grid_search
```

```python
best_forest = RandomForestClassifier(n_estimators = best_n_estimators, max_depth=best_depth, random_state=10)
best_forest.fit(X_train, y_train)
```

## Сравнить качество всех моделей на обучающей и тестовой выборке отдельно по метрикам Accuracy, ROC-AUC, Precision, Recall, F1-мера

```python
from sklearn.metrics import accuracy_score, roc_auc_score, precision_score, recall_score, f1_score, precision_recall_curve
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

models = {
    "K-Nearest Neighbors": best_knn,
    "Logistic Regression": best_log_reg,
    "Decision Tree": best_dec_tree,
    "Random Forest": best_forest
}

results = []

for model_name, model in models.items():
    model.fit(X_train, y_train)
    
    y_train_prob = model.predict_proba(X_train)[:, 1]
    y_test_prob = model.predict_proba(X_test)[:, 1]

    precision, recall, thresholds = precision_recall_curve(y_train, y_train_prob)
    print(model_name, thresholds)
    diff = np.abs(precision - recall)
    best_index = np.argmin(diff)
    best_threshold = thresholds[best_index]
    print(f'Лучший порог отсечения: {best_threshold}')
    plt.plot(thresholds, precision[:-1], label='Precision')
    plt.plot(thresholds, recall[:-1], label='Recall')
    plt.xlabel('Порог отсечения')
    plt.ylabel('Метрики')
    plt.title(f'{model_name}')
    plt.legend()
    plt.show()
    
    threshold = best_threshold 
    if model_name == 'Decision Tree': threshold = 0.7
    y_train_pred = (y_train_prob >= threshold).astype(int)
    y_test_pred = (y_test_prob >= threshold).astype(int)
    
    cm = confusion_matrix(y_test, y_test_pred)

    # Визуализация матрицы ошибок
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Отрицательный (0)', 'Положительный (1)'], yticklabels=['Отрицательный (0)', 'Положительный (1)'])
    plt.xlabel('Предсказанные значения')
    plt.ylabel('Истинные значения')
    plt.title('Матрица ошибок')
    plt.show()

    metrics = {
        "Model": model_name,
        "threshold": threshold,
        "Accuracy (Train)": accuracy_score(y_train, y_train_pred),
        "Accuracy (Test)": accuracy_score(y_test, y_test_pred),
        "Precision (Train)": precision_score(y_train, y_train_pred),
        "Precision (Test)": precision_score(y_test, y_test_pred),
        "Recall (Train)": recall_score(y_train, y_train_pred),
        "Recall (Test)": recall_score(y_test, y_test_pred),
        "F1 Score (Train)": f1_score(y_train, y_train_pred),
        "F1 Score (Test)": f1_score(y_test, y_test_pred),
        "ROC-AUC (Train)": roc_auc_score(y_train, y_train_prob),
        "ROC-AUC (Test)": roc_auc_score(y_test, y_test_prob)
    }
    
    results.append(metrics)

results_df = pd.DataFrame(results)
results_df
```
